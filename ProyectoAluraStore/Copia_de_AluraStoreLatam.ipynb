```python
import matplotlib.pyplot as plt
import pandas as pd

# Re-importing data to ensure DataFrames are defined
url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1_ingresos = tienda["Precio"].sum()
tienda2_ingresos = tienda2["Precio"].sum()
tienda3_ingresos = tienda3["Precio"].sum()
tienda4_ingresos = tienda4["Precio"].sum()

ingresos = [tienda1_ingresos, tienda2_ingresos, tienda3_ingresos, tienda4_ingresos]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ingresos, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
plt.xlabel('Tiendas')
plt.ylabel('Ingresos Totales')
plt.title('Ingresos Totales por Tienda')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# 2. Ventas por categoría

```python
# 1- dataset.groupby Agrupa el DataFrame por la columna .
# 2- ['Producto'].count() Cuenta cuántos productos hay en cada categoría.
# 3- sort_values(ascending=False)	Ordena los resultados de mayor a menor cantidad de productos por categoría.
def agrupar_categoria(dataset):
  return dataset.groupby('Categoría del Producto')['Producto'].count().sort_values(ascending=False)
```

```python
tienda1_categorias = agrupar_categoria(tienda)
tienda1_categorias
```

```python
import matplotlib.pyplot as plt
import pandas as pd

tienda1_categorias = agrupar_categoria(tienda)
etiquetas = ['Muebles', 'Electronicos', 'Jueguetes', 'Electrodomesticos', 'Deportes', 'Musica', 'Libros', 'Hogar']
colores = ['orange', 'green', 'steelblue', 'yellow', 'pink', 'brown', 'palegreen', 'purple']

plt.figure(figsize=(5, 5))
plt.pie(tienda1_categorias, labels=etiquetas, colors=colores, autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Categorías de Productos en Tienda 1')
plt.show()
```

Objetivo: Identificar la tienda con desempeño más débil en los cinco rubros clave para su venta estratégica y seleccionar la tienda más adecuada para ser vendida, maximizando el valor de venta y considerando rendimiento operativo.


## **Análisis Estratégico**
• Ingresos: La Tienda 1 lidera con el mayor ingreso total, lo que la hace atractiva desde el punto de vista financiero.

• Calificación de Clientes: La Tienda 3 tiene la mejor reputación con una calificación promedio de 4.05, lo que indica alta satisfacción del cliente.

• Costo de Envío: La Tienda 4 tiene el costo de envío más bajo, lo que podría traducirse en mayor margen operativo, aunque sus ingresos son los más bajos.

## **Evaluación de Bajo Rendimiento**
• Menor ingreso total: Tienda 4

• Menor calificación promedio: Tienda 1

• Mayor costo de envío: Tienda 1

Tienda 4 tiene el menor ingreso, pero la tienda 1 presenta debilidades en s rubros estrategicos como: calificación y costo de envío.

## **Decisión Final**

Tienda Seleccionada para Venta: Tienda 1

## **Justificacion**
• La tienda 1 tiene mayor ingreso acumulado, lo que eleva su valor de mercado y atractivo para compradores.

• Aunque su calificación es ligeramente inferior a la de Tienda 2 y 3, sigue siendo competitiva.

• El costo de envío es el más alto, pero esto puede ser optimizado por el comprador, mientras que los ingresos y la base de clientes ya están consolidados.

• Aunque Tienda 4 tiene el menor ingreso, Tienda 1 combina baja calificación y alto costo de envío, lo que indica menor eficiencia operativa y menor satisfacción del cliente.

• Vender Tienda 1 permite al dueño deshacerse de una operación costosa y menos valorada por los clientes, liberando capital para un nuevo negocio con mejor potencial.

## **Recomendación al Dueño**
Se recomienda vender la tienda 1 por su alto valor financiero, lo que permitirá obtener el mejor retorno económico. Las otras tiendas pueden seguir operando con ajustes estratégicos para mejorar ingresos o eficiencia.
